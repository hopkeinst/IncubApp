```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://www.programmableweb.com/category/tools/api"

cnt_apis = 0
apis_dict = {}

while True:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, "html.parser")
    apis_even = soup.find_all("tr", {"class": "even"})
    apis_odd = soup.find_all("tr", {"class": "odd"})
    
    for api in apis_even:
        cnt_apis += 1
        title = api.find("td", {"class": "views-field-title"}).text[1:]
        link = "https://www.programmableweb.com" + api.find("td", {"class": "views-field-title"}).find("a").get("href")
        category = api.find("td", {"class": "views-field-field-article-primary-category"}).text[1:]
        description = api.find("td", {"class": "views-field-field-api-description"}).text[1:]
        apis_dict[cnt_apis] = [title, link, category, description]
        
    for api in apis_odd:
        cnt_apis += 1
        title = api.find("td", {"class": "views-field-title"}).text[1:]
        link = "https://www.programmableweb.com" + api.find("td", {"class": "views-field-title"}).find("a").get("href")
        category = api.find("td", {"class": "views-field-field-article-primary-category"}).text[1:]
        description = api.find("td", {"class": "views-field-field-api-description"}).text[1:]
        apis_dict[cnt_apis] = [title, link, category, description]
        
    url_tag = soup.find("a", {"id": "pager_id_apis_all"})
    if url_tag != None:
        url = "https://www.programmableweb.com" + url_tag.get("href")
    else:
        break
    
print("Total apis:", cnt_apis)
```

```python
apis_df = pd.DataFrame.from_dict(apis_dict, orient="index", columns=["API Name", "API URL", "API Category", "API Description"])
apis_df.to_csv("apis_tools.csv")
```
