## 06 Web Scraping Craiglist - Job Description Page

```python
for job in jobs:
    title = (job.find("a", {"class": "result-title"})).text
    location_tag = job.find("span", {"class": "result-hood"})
    location = location_tag.text[2:-1] if location_tag else "N/A"
    date = job.find("time", {"class":"result-date"}).get("title")
    link = job.find("a", {"class":"result-title"}).get("href")
    
    job_response = requests.get(link)
    job_data = job_response.text
    job_soup = BeautifulSoup(job_data, "html.parser")
    job_description = job_soup.find("section", {"id":"postingbody"}).text
    job_attributes_tag = job_soup.find("p", {"class":"attrgroup"})
    job_attributes = job_attributes_tag.text if job_attributes_tag else "N/A"
    
    print("Titulo:", title, "\n\tUbicacion:", location, "\n\tFecha:", date, "\n\tWeb:", link, "\n\tDescripcion:", job_description, "\n\tAtributos:", job_attributes, "\n-----")
```

## 07 Crawling & Scraping Next Pages

```python
url_nextP = "https://boston.craigslist.org/search/etc"
cnt_jobs = 0

while True:
    response = requests.get(url_nextP)
    data = response.text
    soup = BeautifulSoup(data, "html.parser")
    jobs = soup.find_all("div", {"class":"result-info"})
    
    for job in jobs:
        cnt_jobs += 1
        title = (job.find("a", {"class": "result-title"})).text
        location_tag = job.find("span", {"class": "result-hood"})
        location = location_tag.text[2:-1] if location_tag else "N/A"
        date = job.find("time", {"class":"result-date"}).get("title")
        link = job.find("a", {"class":"result-title"}).get("href")

        job_response = requests.get(link)
        job_data = job_response.text
        job_soup = BeautifulSoup(job_data, "html.parser")
        job_description = job_soup.find("section", {"id":"postingbody"}).text
        job_attributes_tag = job_soup.find("p", {"class":"attrgroup"})
        job_attributes = job_attributes_tag.text if job_attributes_tag else "N/A"

        print("#"*86)
        print("##########" + "{:^66s}".format("TRABAJO # "+str(cnt_jobs)) + "##########")
        print("#"*86)
        print("Titulo:", title, "\n\tUbicacion:", location, "\n\tFecha:", date, "\n\tWeb:", link, "\n\tDescripcion:", job_description, "\n\tAtributos:", job_attributes)
        print("#"*86)
        print()
    
    url_tag = soup.find("a", {"title": "next page"})
    if url_tag.get("href"):
        url_nextP = "https://boston.craigslist.org" + url_tag.get("href")
        print("### NUEVA URL:", url_nextP)
    else:
        break
        
print("#"*86)
print("Total Trabajos: "+str(cnt_jobs))
print("#"*86)
        
```
